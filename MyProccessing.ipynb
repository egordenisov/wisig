```python
import pickle
import numpy as np
import matplotlib.pyplot as plt


with open('dataset/TwoRxOneTxSingleDay.pkl', 'rb') as f:
    d = pickle.load(f)
    
capture_date_list=d['capture_date_list']
tx_list= d['tx_list']
rx_list= d['rx_list']
data = d['data']

print(capture_date_list)
print(tx_list)
print(rx_list)
print(np.shape(data))
```

'data' is 7 dimentional array
[tx rx, days, eq, signals, points, iq]

```python
raw_dsp = data[0][0][0][0][200]
si = raw_dsp[:,0]
sq = raw_dsp[:,1]

plt.plot(si)
plt.plot(si*si + sq*sq)
```

```python
eq = 0
train_data = data[0][0][0][eq][:,:,0] + 1j * data[0][0][0][eq][:,:,1]
test_data =  data[0][1][0][eq][:,:,0] + 1j * data[0][1][0][eq][:,:,1]
print(np.shape(train_data))
print(np.shape(test_data))
```
